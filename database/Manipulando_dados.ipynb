```python
#Adicionando tipo de temperatura
paises_quentes = [
    'Algeria', 'Bahamas', 'Bahrain', 'Bangladesh', 'Barbados', 'Botswana',
    'Cambodia', 'Cameroon', 'Colombia', 'Costa Rica', 'Cuba',
    'Dominican Republic', 'Egypt', 'El Salvador', 'Fiji', 'Ghana', 'Guatemala',
    'Hong Kong (China)', 'India', 'Indonesia', 'Iraq', 'Israel', 'Jamaica',
    'Jordan', 'Kenya', 'Kuwait', 'Libya', 'Madagascar', 'Malaysia', 'Mauritius',
    'Morocco', 'Nigeria', 'Oman', 'Panama', 'Paraguay', 'Philippines', 'Qatar',
    'Saudi Arabia', 'Singapore', 'Sri Lanka', 'Syria', 'Taiwan', 'Tanzania',
    'Thailand', 'Trinidad And Tobago', 'Tunisia', 'United Arab Emirates',
    'Uruguay', 'Venezuela', 'Vietnam', 'Zimbabwe'
]


paises_frios = [
    'Armenia', 'Austria', 'Belarus', 'Belgium', 'Canada', 'Czech Republic',
    'Denmark', 'Estonia', 'Finland', 'Germany', 'Iceland', 'Ireland', 'Kazakhstan',
    'Latvia', 'Lithuania', 'Luxembourg', 'Moldova', 'Netherlands', 'Norway',
    'Poland', 'Russia', 'Slovakia', 'Slovenia', 'Sweden', 'Switzerland', 'Ukraine',
    'United Kingdom', 'Uzbekistan'
]

paises_mistos = [
    'Albania', 'Argentina', 'Australia', 'Bolivia', 'Bosnia And Herzegovina',
    'Brazil', 'Bulgaria', 'Chile', 'China', 'Croatia', 'Cyprus', 'France', 'Georgia',
    'Greece', 'Hungary', 'Iran', 'Italy', 'Japan', 'Lebanon', 'Mexico',
    'Montenegro', 'Nepal', 'New Zealand', 'Pakistan', 'Peru', 'Portugal',
    'Romania', 'Serbia', 'South Africa', 'South Korea', 'Spain', 'Turkey',
    'United States'
]

def classificar_temperatura (pais):
    if pais in paises_quentes:
        return 'Quente'
    elif pais in paises_frios:
        return 'Frio'
    elif pais in paises_mistos:
        return 'Variavel'
    else:
        return 'Indefinido'

df['Temperatura'] = df['Country'].apply(classificar_temperatura)
```

```python
#Adicionando tipo de umidade
paises_umidos = [
    'Bangladesh', 'Barbados', 'Cambodia', 'Colombia', 'Costa Rica',
    'Cuba', 'Dominican Republic', 'Fiji', 'Ghana', 'Guatemala', 'Hong Kong (China)',
    'India', 'Indonesia', 'Jamaica', 'Malaysia', 'Mauritius', 'Nigeria',
    'Panama', 'Paraguay', 'Philippines', 'Singapore', 'Sri Lanka', 'Taiwan',
    'Thailand', 'Trinidad And Tobago', 'Vietnam'
]

paises_secos = [
    'Algeria', 'Australia', 'Bahrain', 'Botswana', 'Egypt', 'Iran', 'Iraq',
    'Israel', 'Jordan', 'Kuwait', 'Libya', 'Morocco', 'Oman', 'Qatar', 'Saudi Arabia',
    'Syria', 'Tunisia', 'United Arab Emirates'
]

paises_mistos_clima = [
    'Argentina', 'Armenia', 'Austria', 'Belarus', 'Bolivia', 'Bosnia And Herzegovina','Brazil',
    'Bulgaria', 'Canada', 'Chile', 'China', 'Croatia', 'Cyprus', 'Denmark', 'Estonia',
    'France', 'Georgia', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Italy', 'Japan',
    'Kazakhstan', 'Lebanon', 'Mexico', 'Montenegro', 'Nepal', 'New Zealand', 'Norway',
    'Pakistan', 'Peru', 'Portugal', 'Romania', 'Russia', 'Serbia', 'South Africa',
    'South Korea', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'Ukraine',
    'United Kingdom', 'United States', 'Uzbekistan', 'Venezuela', 'Zimbabwe'
]

def classificador_umidade (pais):
    if pais in paises_umidos:
        return 'Umido'
    elif pais in paises_secos:
        return 'Seco'
    elif pais in paises_mistos_clima:
        return 'Misto'
    else:
        return 'Indeterminado'

df['Umidade'] = df['Country'].apply(classificador_umidade)
```

```python
#Adicionando tipo de paisagem
paises_montanhosos = [
    'Armenia', 'Austria', 'Bolivia', 'Bosnia And Herzegovina', 'Chile', 'China',
    'Croatia', 'Ecuador', 'Ethiopia', 'Georgia', 'Greece', 'Iceland', 'India',
    'Iran', 'Italy', 'Japan', 'Kazakhstan', 'Lebanon', 'Mexico', 'Montenegro',
    'Morocco', 'Nepal', 'New Zealand', 'Norway', 'Pakistan', 'Peru', 'Portugal',
    'Romania', 'Serbia', 'South Korea', 'Spain', 'Switzerland', 'Turkey',
    'Ukraine'
]

paises_florestais = [
    'Bangladesh', 'Brazil', 'Canada', 'Colombia', 'Costa Rica', 'Cuba', 'Finland',
    'France', 'Germany', 'Guatemala', 'Indonesia', 'Malaysia', 'Panama', 'Paraguay',
    'Peru', 'Philippines', 'Russia', 'Sweden', 'Thailand', 'Trinidad And Tobago',
    'United States', 'Venezuela', 'Vietnam'
]

paises_praianos = [
    'Australia', 'Bahamas', 'Barbados', 'Dominican Republic', 'Fiji', 'Greece',
    'Honduras', 'Israel', 'Italy', 'Jamaica', 'Maldives', 'Mauritius', 'Mexico',
    'New Zealand', 'Portugal', 'Sri Lanka', 'Thailand', 'Trinidad And Tobago'
]

paises_urbanos = [
    'Bahrain', 'Hong Kong (China)', 'Japan', 'Singapore', 'South Korea',
    'United Arab Emirates'
]


def paisagem (pais):
    if pais in paises_montanhosos:
        return 'Montanhosa'
    elif pais in paises_florestais:
        return 'Florestal'
    elif pais in paises_praianos:
        return 'Praiano'
    elif pais in paises_urbanos:
        return 'Urbano'
    else:
        return 'Indefinido'

df['Paisagem'] = df['Country'].apply(paisagem)
```

```python
#Adicionando nível de segurança
paises_baixa_seguranca = [
    'Afghanistan', 'Central African Republic', 'Haiti', 'Iraq', 'Libya', 'Mali', 
    'Somalia', 'South Sudan', 'Syria', 'Venezuela', 'Zimbabwe', 'Nigeria', 
    'Pakistan', 'Yemen', 'Bangladesh', 'North Korea', 'Palestine'
]

paises_media_seguranca = [
    'Albania', 'Algeria', 'Argentina', 'Armenia', 'Azerbaijan', 'Belarus', 
    'Bolivia', 'Bosnia And Herzegovina', 'Brazil', 'Cambodia', 'Cameroon', 
    'Colombia', 'Cuba', 'Dominican Republic', 'Ecuador', 'Egypt', 'El Salvador', 
    'Georgia', 'Guatemala', 'Honduras', 'India', 'Indonesia', 'Iran', 'Israel', 
    'Jamaica', 'Jordan', 'Kazakhstan', 'Kenya', 'Kosovo (Disputed Territory)', 
    'Kyrgyzstan', 'Lebanon', 'Madagascar', 'Malaysia', 'Mexico', 'Moldova', 
    'Morocco', 'Nepal', 'Paraguay', 'Peru', 'Philippines', 'Russia', 'Serbia', 
    'South Africa', 'Sri Lanka', 'Thailand', 'Turkey', 'Ukraine', 'Uzbekistan', 
    'Vietnam'
]

paises_alta_seguranca = [
    'Australia', 'Austria', 'Bahamas', 'Bahrain', 'Barbados', 'Belgium', 
    'Botswana', 'Canada', 'Chile', 'China', 'Croatia', 'Cyprus', 'Czech Republic', 
    'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hong Kong (China)', 
    'Hungary', 'Iceland', 'Ireland', 'Italy', 'Japan', 'Kuwait', 'Latvia', 
    'Lithuania', 'Luxembourg', 'Malta', 'Mauritius', 'Montenegro', 'Netherlands', 
    'New Zealand', 'Norway', 'Panama', 'Poland', 'Portugal', 'Puerto Rico', 'Qatar', 
    'Saudi Arabia', 'Singapore', 'Slovakia', 'Slovenia', 'South Korea', 'Spain', 
    'Sweden', 'Switzerland', 'Taiwan', 'Trinidad And Tobago', 'United Arab Emirates', 
    'United Kingdom', 'United States', 'Uruguay'
]

def seguranca (pais):
    if pais in paises_baixa_seguranca:
        return 'Baixa'
    elif pais in paises_media_seguranca:
        return 'Media'
    elif pais in paises_alta_seguranca:
        return 'Alta'
    else:
        return 'Inderteminado'

df['Nível_segurança'] = df['Country'].apply(seguranca)
```
